STANDARISASI DATA DAN KORELASI DATA

```python
def standarisasi_data():
    # Standarisasi format Gender
    df['Gender'] = df['Gender'].replace({'P': 'Perempuan', 'L': 'Laki-laki'})

    # --- Standarisasi format kolom 'Diagnose' ---
    df['Diagnose'] = df['Diagnose'].astype(str).str.strip().str.upper()  # hilangkan spasi dan jadikan huruf besar

    # Encoding kategorikal
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    df['Gender'] = le.fit_transform(df['Gender'])
    df['Diagnose'] = le.fit_transform(df['Diagnose'])

    print(df.head())
    return df.head()
```
