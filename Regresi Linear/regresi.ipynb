```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data X dan Y dari soal
data = {
    "X": [20, 40, 20, 30, 27, 42, 20, 21, 29, 32, 43, 38],
    "Y": [78, 150, 81, 115, 102, 152, 82, 83, 110, 70, 160, 135]
}

df = pd.DataFrame(data)

# Hitung rata-rata
x_mean = df["X"].mean()
y_mean = df["Y"].mean()

# Hitung kolom bantu
df["X_Xbar"] = df["X"] - x_mean
df["Y_Ybar"] = df["Y"] - y_mean
df["XY_diff"] = df["X_Xbar"] * df["Y_Ybar"]
df["X_diff_sq"] = df["X_Xbar"] ** 2
df["Y_diff_sq"] = df["Y_Ybar"] ** 2

# Hitung b dan a
b = df["XY_diff"].sum() / df["X_diff_sq"].sum()
a = y_mean - b * x_mean

print(f"Rata-rata X (x̄): {x_mean}")
print(f"Rata-rata Y (ȳ): {y_mean}")
print(f"Koefisien b (slope): {b}")
print(f"Intercept a: {a}")
print(f"Persamaan regresi: Y = {a:.2f} + {b:.2f}X")

# Prediksi jika X = 25
x_pred = 45
y_pred = a + b * x_pred
print(f"Prediksi Y jika X = 45: {y_pred:.2f}")

# Koefisien korelasi r dan determinasi r^2
r = df["XY_diff"].sum() / np.sqrt(df["X_diff_sq"].sum() * df["Y_diff_sq"].sum())
r_squared = r ** 2
print(f"Koefisien korelasi (r): {r:.3f}")
print(f"Koefisien determinasi (r²): {r_squared:.3f} ({r_squared*100:.2f}%)")

# Visualisasi
plt.scatter(df["X"], df["Y"], color="blue", label="Data Asli")
plt.plot(df["X"], a + b * df["X"], color="red", label="Regresi Linear")
plt.xlabel("Waktu Treadmill (menit)")
plt.ylabel("Kalori Terbakar")
plt.title("Regresi Linear Sederhana")
plt.legend()
plt.grid(True)
plt.show()
```
