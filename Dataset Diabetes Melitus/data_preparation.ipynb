```python
import pandas as pd

import import_ipynb
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.feature_selection import SelectKBest, f_classif
from manipulate_data import show_data
```

AMBIL DATA

```python
path = 'Dataset.xlsx'
df = show_data(path)
```

HAPUS DATA DUPLIKAT

```python
def duplicate_data():
    hapus_duplikat = df.drop_duplicates()
    print(hapus_duplikat)
    return hapus_duplikat

duplicate_data()
```

STANDARISASI DATA DAN KORELASI DATA

```python
def standarisasi_data():
    # Standarisasi format Gender
    df['Gender'] = df['Gender'].replace({'P': 'Perempuan', 'L': 'Laki-laki'})
    
    print(df)

    return df

print(standarisasi_data())
```

```python
def encoding_data():

    
    le = LabelEncoder()
    df['Gender'] = le.fit_transform(df['Gender'])
    df['Diagnose'] = le.fit_transform(df['Diagnose'])

    print(df)
    return df

print(encoding_data())
```

NORMALISASI NUMERIK

```python
def normalisasi_data():

    numeric_cols = ['Age', 'Hemoglobin', 'Hematokrit', 'Lekosit', 
                    'Eritrosit', 'Trombosit', 'HbA1c', 'RBG']

    minmax_scaler = MinMaxScaler()
    df_minmax = df.copy()
    df_minmax[numeric_cols] = minmax_scaler.fit_transform(df[numeric_cols])

    return df_minmax

    
normalisasi_data()
```

Reduksi Dimensi

```python

def reduksi_data():
    df =  normalisasi_data()

    x = df[['Age', 'Hemoglobin', 'Hematokrit', 'Lekosit', 'Eritrosit', 'Trombosit', 'HbA1c', 'RBG']]
    y = df['Diagnose']

    # Seleksi 3 fitur terbaik berdasarkan nilai F
    selector = SelectKBest(score_func=f_classif, k=3)
    X_new = selector.fit_transform(x, y)

    selected_features = x.columns[selector.get_support()]
    print("✅ Fitur terbaik berdasarkan ANOVA F-test:" , selected_features)
    return selected_features

reduksi_data()
```

AKSIKAN FUNGSI DATA PREPARASI

```python
def data_preparation():
    return duplicate_data(), standarisasi_data(), encoding_data(), normalisasi_data(), reduksi_data()

data_preparation()
```
