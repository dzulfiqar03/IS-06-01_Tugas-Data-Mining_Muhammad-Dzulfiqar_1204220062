```python
path = 'Dataset.xlsx'
show_data(path)


def save_all():

    print_data()
    deskripsi, isnull_df, df = understanding()
    duplikasi = data_preparation()
    with pd.ExcelWriter('Dataset_Bersih.xlsx', engine='openpyxl', mode='w') as writer:
            # Sheet utama
        df.to_excel(writer, sheet_name='Data Utama', index=False)
        # Sheet deskripsi
        deskripsi.to_excel(writer, sheet_name='Deskripsi')
        # Sheet null summary
        isnull_df.to_excel(writer, sheet_name='isNull')
        duplikasi.to_excel(writer, sheet_name='Duplikasi')
    
save_all()
```
